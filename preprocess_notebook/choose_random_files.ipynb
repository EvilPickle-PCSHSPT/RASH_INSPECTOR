# Import Libraries

```python
import os
import random
import shutil
```

# Enter the Folder Path

```python
src_path = '/Users/waranthornchansawang/Desktop/RASH_INSPECTOR/Dataset/ORGAN/step5_augmented_data/arm/AD'
new_path = '/Users/waranthornchansawang/Desktop/RASH_INSPECTOR/Dataset/ORGAN/step6_random_data/arm/AD'
```

# Enter Number of Files that You Want

```python
filenames = random.sample(os.listdir(src_path), 300)
for files in filenames:
    srcpath = os.path.join(src_path, files)
    #print(srcpath)
    shutil.copy(srcpath, new_path)
```
